## Introduction
Every human oprates in a system of global economy which include: production, consumption, economic management, work in general, exchange of financial values and trade of goods and services within and outside their countries. The aim of this project is to review our collective performances over the years seeing that we have come a long way


### Dataset Description 
The data we would be using to explore how the world economy has performed over the years is a public dataset and it is available for dowload on  [Gap Minder](https://gapminder.org/data/). However, there was no comprehensive data detailing the countries and their continents on [Gap Minder](https://gapminder.org/data/) which was why the `continents'` dataset was sourced from [Our World in Data](https://ourworldindata.org/grapher/continents-according-to-our-world-in-data)

### Data Dictionary

- ` country`: The countries of the world

- `continent`: The continents of the world

- `year` : The year for each record

- `agro`: Expressed as a percentage of GDP. It respresents the value of agricultural products delivered to the world

- `import`:Expressed as a percentage of GDP and it respresents the value of goods and other market services recieved from the rest of the world

- `Export`: Expressed as a percentage of GDP and it respresents the value of goods and other market services delivered to the rest of the world

- `investment`:Expressed as a percentage of GDP and it respresents the value financial services delivered to the rest of the world

- `industry`: Expressed as a percentage of GDP and it respresents the value of industrial products delivered to the world

- `Services`: Expressed as a percentage of GDP and it respresents the value Services delivered to the world

- `gdp`: Gross Domestic Product converted to international Dollar using power parity rate

__N:B__ The above listed columns are contained in cleaned dataset

### Questions for Analysis
1. How has the world economy performed over the years?
2. What is the general trend displayed by all continents?

```python
#import all the neccessary packages and styles for the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('default')
```

```python
#create list of all the csv files to be read into pandas
dataset=[
    "agriculture_percent_of_gdp.csv",
    "continents-according-to-our-world-in-data.csv",
    "exports_percent_of_gdp.csv",
    "imports_percent_of_gdp.csv",
    "industry_percent_of_gdp.csv",
    "investments_percent_of_gdp.csv",
    "services_percent_of_gdp.csv",
    "total_gdp_ppp_inflation_adjusted.csv"
]
```

```python
# the function reads csv files according to its index
def load(data,num):
    """reads all the csv files"""
    return pd.read_csv(data[num])


#Load all the dataset
agric_df=load(dataset,0) 
geo_df=load(dataset,1)   
export_df=load(dataset,2)
import_df=load(dataset,3) 
industry_df=load(dataset,4) 
investment_df=load(dataset,5) 
services_df=load(dataset,6) 
gdp_df=load(dataset,7) 

agric_df.head(2)
```

### The need to transpose
Majority of all the datset are in wide format as shown above. Hence the need to transpose them. In pandas, the melt() function does the trick. 

```python
#create a function that performs the melts the operation 
def transpose(df,var):
    df=pd.melt(df, id_vars='country', var_name='year', value_name=var)
    return(df)

#*********************************************************************#
#melt the dataset accordingly
agric_df=transpose(agric_df,"agro")
export_df=transpose(export_df,"export")
import_df=transpose(import_df,"import")
industry_df=transpose(industry_df,"industry")
investment_df=transpose(investment_df,'investment')
services_df=transpose(services_df,"services")
gdp_df=transpose(gdp_df,"gdp")
gdp_df.tail()
```

```python
geo_df.head(2)#preview dataset
```

### Merge all dataset together
All the data read into pandas will be merged into a single dataframe using "inner join."

   __similarities in the dataset__
- `geo_df` has column "Entity" which is equivalent to 'country' column in `agric_df`.
- The others have "country" and "year" in common as columns

The `geo_df` and `agric_df` will be merged first and separately. The resulting dataframe will then be merged with others

```python
#use inner join to merg agro_df and geo_df 
df=pd.merge(agric_df,geo_df[["Entity","Continent"]],right_on="Entity",left_on="country").drop(columns=["Entity"])
data_set=[import_df,export_df,industry_df,investment_df,services_df,gdp_df]
```

```python
#create a function that merges all other dataset at once
def merge(df,others):
    for i,v in enumerate(others):
        df=pd.merge(df,others[i],on=['country','year'])
    return df

merged_df=merge(df,data_set)
merged_df.head()
```

## Data Wrangling

At a glance the data looks too dirty. There are lots of empty entries which are not useful in the analysis. Their presence will only cause more harm than good. Hence, the need to sort out the NaN and subsequently get the data to be clean
 
At this point, it is necessary to stress that the GDP, `"gdp" column ` is the dependent variable. Any empty cell there will not be tolerated. As such, they will be dropped.

Likewise, all the rows in which all the following variables,`agro, import, export, investment, services` are empty will be dropped.




### Data Cleaning

```python
#
merged_df.info()
```

```python
#drop a rows with the 5 columns empty at a time
merged_data=merged_df.dropna(axis=0,how="all",subset=["agro","import","export","investment","services"]).reset_index(drop=True)
merged_data.rename(columns={"Continent":"continent"},inplace=True)
merged_data.head()
```

### Inspecting data before further cleaning
As previewed above, The empty cells seems to have all gone. It is a big fat __NO__. Just the first few rows of the entire data that were previewed. The data will be subjected to further inspection to know what next to clean. However, we can see that the `gdp` columns contains characters. the B means "billion". Also, the `contentent` and `region` were not entered appropriately. 


__UP NEXT,__
             We make changes to the data based on the following observations:
             
 - `Continent` should be in sentence case
 - `regions` should in sentence case and drop irrelevant strings
 - `dgp` contains characters that may prevent string conversion to float
             
             
       


```python
merged_df=merged_data.copy()# make a copy of the dataset before further cleaning
merged_df['continent']=merged_df['continent'].str.title()#convert column to sentence case
merged_df['continent'].value_counts()#check if changes have been made
```

### Cleaning GDP column
Figures are designated in Millions, Billions, and Trillions as the case may be.Hence need to extract and multiply `dgp` appropriately

```python
#target the Characters in the GDP column and split it to other columns
merged_df[["gdp_fig","unit_char","empty_cells"]]=merged_df['gdp'].str.split(r'([a-zA-Z]+)',expand=True,n=1)
merged_df.head()
```

```python
# B-Billion,M-Million,TR-Trillion
dic = {
    'B': '1000000000',
    'TR': '1000000000000',
    'M': '1000000'
}

merged_df['gdp_format'] = merged_df['unit_char'].replace(dic).astype(float)
merged_df['gdp_format']=merged_df['gdp_fig'].astype(float)*merged_df['gdp_format']
merged_df.drop(columns=["gdp","gdp_fig","unit_char","empty_cells"],inplace=True)
```

### data cleaning bothering on outliers, data types and null values(NaN)
1. All numeric data in our data are expressed in percentage except `gdp` and `year`. As such, any value __>=100__ is not realistic. Besides, no country has been reported to have achieved such milestone in many years.

2. check all data types

3. Fill NaN with the mean of their their respective countries.

4. And lastly, any NaN after step 3 will be dropped eventually


```python
#get snapshot of the data
merged_df.describe()
```

```python
#filter out all export and import observations greater than or equal to 100%
filtered_import=merged_df[merged_df['import']<100]
filtered_df=filtered_import[filtered_import['export']<100]
fifiltered_df=filtered_import[filtered_import['export']<100]
new_df=filtered_df.reset_index(drop=True)
new_df.describe()
```

### Change data types
The `year` and `investment` are not in the right format. They would be converted to type `int` and `float` respectively.
Howver, the  `investment` column has a character that will hinder the conversion which must be removed or replaced

```python
new_df['year']=new_df['year'].astype(int) #convert to type int
new_df['investment'] = new_df['investment'].str.replace('âˆ’', '-') #replace 

new_df['investment']=new_df['investment'].astype(float)
new_df=new_df.dropna(axis=0,how="all",subset=["gdp_format"]).reset_index()
new_df.dtypes
```

### Fill NA with the means of their respective countries for every column that NaN is present

```python
#create a function that fills NaN with mean per country of any numeric column
def fil(df,col):
    dict=df.groupby(["country"])[str(col)].mean().dropna().to_dict()
    df.loc[df[str(col)].isnull(), str(col)] = new_df['country'].map(dict)
    return df

#stepwisely fill NaN across the columns
fil_agro_na=fil(new_df,'agro') 
fil_industry=fil(fil_agro_na,'industry')
fil_investment=fil(fil_industry,'investment')
fil_services=fil(fil_investment,"services")

#drop all other null values
cleaned_data=fil_services.dropna().drop(labels='index',axis=1)

cleaned_data.info()
```

## How has the world economy performed since 1960?

```python
sum_cont=df.groupby(["continent",'year'],as_index=False)['gdp_format'].sum() #sum of GDP per continent each year
sum_cont["gdp_in_trillions"]=(sum_cont["gdp_format"]/(10**12))
continent=sum_cont.groupby('continent',as_index=False)['gdp_in_trillions'].mean().sort_values('gdp_in_trillions',ascending=False)
continent
```

```python
explode = (0.15, 0, 0, 0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(continent['gdp_in_trillions'], explode=explode, labels=continent['continent'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title(" Average GDP of the world since 1960 is {} Trillion".format(round(sum(continent["gdp_in_trillions"]),2)),
              fontsize=20)

plt.show()
```

>### Findings from the chart
Bewteen 1960-2013, the world's average GDP was 41.62 Trillion and Asia presented as the largest continent among others accounting for 35.2% of world's mean GDP. This align with a more recent study carried out by [statistics Times](https://statisticstimes.com/economy/continents-by-gdp.php#:~:text=Asia%20is%20the%20largest%20continental,%2468.7%20trillion.).  Oceania on the other hand comprised 1.3% followed by Africa with 6.7%, while Europe and North America's GDP combined is just over half of GDP.
It is important to stress that this is the weighted avrerage of the gross and did not speak for a fact of the 
underlying trends. It would be neccesary to get into more details to know what played out throughout the period. 


```python
cont_df=df.groupby(["country",'continent',"year"],as_index=False)["gdp_format"].sum()
cont_df["gdp_in_trillions"]=cont_df['gdp_format']/(10**12)
cont_df
```

# What is the growth like for every continents since 1960?

```python
plt.style.use('fivethirtyeight')
df["gdp_in_trillions"]=df['gdp_format']/(10**12)
asia=df.query("continent=='Asia'").groupby("year",as_index=False)["gdp_in_trillions"].sum()
s_america=df.query("continent=='South America'").groupby("year",as_index=False)["gdp_in_trillions"].sum()
n_america=df.query("continent=='North America'").groupby("year",as_index=False)["gdp_in_trillions"].sum()
europe=df.query("continent=='Europe'").groupby("year",as_index=False)["gdp_in_trillions"].sum()
africa=df.query("continent=='Africa'").groupby("year",as_index=False)["gdp_in_trillions"].sum()
oceania=df.query("continent=='Oceania'").groupby("year",as_index=False)["gdp_in_trillions"].sum()

fig,ax=plt.subplots(figsize=(8,6))
ax.plot(asia.year,asia["gdp_in_trillions"],label="Asia", linewidth=4)
ax.plot(s_america.year,s_america["gdp_in_trillions"],label="South America", linewidth=4)
ax.plot(n_america.year,n_america["gdp_in_trillions"],label="North America", linewidth=4)
ax.plot(europe.year,europe["gdp_in_trillions"],label="Europe", linewidth=4)
ax.plot(africa.year,africa["gdp_in_trillions"],label="Africa", linewidth=4)
ax.plot(oceania.year,oceania["gdp_in_trillions"],label="Oceania", linewidth=4)
ax.set_title("Total GDP PPP inflation adjusted per year")

ax.set_ylabel("GDP (trillions)")

ax.spines["right"].set_visible(False)
ax.spines['top'].set_visible(False)
plt.legend( fontsize=12)
plt.show()
```

>## Findings
Overall, all the continents showed an upward trend in economic growth throughout the period. However, `Asia`, `Europe` and `North America` experienced a surge in GDP growth since 1970 and have made dramatic progress over the years. A more recent study by [statistics Times](https://statisticstimes.com/economy/continents-by-gdp.php#:~:text=Asia%20is%20the%20largest%20continental,%2468.7%20trillion.) reveals similar trends. 

>Between 1970 and 1990,  Asia led North America and Europe marginally. However, in 1990 and beyond, economic growth appereared to be overwhelmingly in support of the `Asian` continent. It will be crucial to find out what played out that Asia displayed sucha remarkable growth in that period. The graph clearly suggests something different about economic activities in Asia compared to the rest of the world.

# How do the continents differ in terms of export, industry and investments?
In this part of the project, certain key economic contributors to the GDP will be invesigated to reveal the economic strength of each continent. They include:`Export`, `Industry`, `Investment` and `Services`. They have been selected because in theory they contribute positively to GDP unlike `Import` that has negative effect on GDP. in other words, the governments spend on every import to their country which has a negative feedback on GDP.

Although `Agriculture` has a positive impact on GDP, it will not be considered in this case because `Agro` is part of a nation's export.

```python
table = df.groupby('continent').mean()
table = table.iloc[:, 3:7].reset_index()
table = table.melt(id_vars='continent', var_name='sector', value_name='percentage')
table.head()
```

```python
plt.style.use('fivethirtyeight')
temp = np.full((6,), 0)
for s in table['sector'].unique():
    sector_data = table.query("sector == @s")
    p=plt.barh(sector_data['continent'], sector_data['percentage'], label=s, left=temp, alpha=0.9)
    plt.bar_label(p,label_type='center',size=8, color='w', fmt= "%.0f", fontweight='bold')
    plt.xticks([])
    plt.yticks(fontsize=9, rotation=15)
    temp += sector_data.percentage.reset_index(drop=True)
    plt.title('What Makes Asia Different?', loc='left', pad=30)
plt.text(x=0, y=5.7, s='Asia appears to have a larger industrial sector compared\nto the rest'+
        ' of the world', fontsize=12, fontstyle='italic')
sns.despine(bottom=True)
    
plt.legend(loc=4, fontsize=9)
plt.show()
```

>### Findings
The above graph clear shows that Asia had heavily invested in their industries while maintaing other sectors in good shape compared to the rest of the the world. In addition, the average Asian investment portfolio is equivalent to that of Europe at 25%. According to [ADB,](https://www.adb.org/publications/asias-industrial-transformation-manufacturing-gvcs-part1) Asian Developmental Bank, revealed in one of their publications that the single most important factor that explains East Asiaâ€™s development success was its fast structural transformation toward industrialization, manufacturing in particular.

> It is best at this point to  narrown down the study down to the countries contributing maximally to the Asian GDP. Let's find out who they are and what they do best.



# Are the top Asian countries  also leaders in Industry?

>#### Probable expectation: Top industrialized countries in Asia are leading in GDP

```python
#filter out every other coninent except Asia
asia = df.query("continent == 'Asia'")
means = asia.groupby('country').mean()
asia_gdp = means.sort_values(by='gdp_in_trillions', ascending=False)
asia_industry = means.sort_values(by='industry', ascending=False)


top_10_gdp = asia_gdp.head(10)['gdp_in_trillions'].sort_values() #top 10 Asian countries leading in GDP
top_10_industry = asia_industry.head(10)['industry'].sort_values()#top 10 Asian countries leading in industry

```

### Commonalities between top the top 10 big players according to GDP and top 10 industry leaders

```python
commonalities=np.intersect1d(top_10_gdp.index,top_10_industry.index)
commonalities
```

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})

fig,ax=plt.subplots(1,2)
p1=ax[0].barh(top_10_gdp.index,top_10_gdp.values,color='#0AA1DD',alpha=0.5)
p2=ax[1].barh(top_10_industry.index,top_10_industry.values,color='#0AA1DD',alpha=0.5)
##ax[0].set_xticks([])
#ax[1].set_xticks([])
ax[0].set_yticks(top_10_gdp.index)
ax[0].set_yticklabels(top_10_gdp.index,rotation=2)
ax[0].set_title("Asian top Countries by GDP",fontweight='bold')
ax[1].set_title("Asian top Industrialised Countries",fontweight='bold')


ax[1].set_yticks(top_10_industry.index)
ax[1].set_yticklabels(top_10_industry.index,rotation=2)

ax[0].set_xlabel("Total GDP in Trillions",fontsize=10)
ax[1].set_xlabel("Fraction of industry (%)",fontsize=10)

p1[9].set_color("#2155CD")
p1[5].set_color("#2155CD")

p2[0].set_color("#2155CD")
p2[4].set_color("#2155CD")



ax[0].grid(False)
ax[1].grid(False)
plt.tight_layout()
sns.despine(bottom=True)
ax[1].grid(alpha=0.2)
ax[0].grid(alpha=0.2)


plt.show()
```
